```python
import numpy as np
import pandas as pd
import joblib
```

```python
data = pd.read_csv("../algebra_2008_2009_train_trimmed3_remove_duplicates_sorted.csv", delimiter='\t')
```

```python
data.describe(include='all')
```

```python
validation = []


previous_idunit = "".join([data['Anon Student Id'][0], data['Unit'][0]])
for i in range(1, len(data)):
    current_idunit = "".join([data['Anon Student Id'][i], data['Unit'][i]])
    if current_idunit != previous_idunit:
        subtractor = 1
        prob = data['Problem Name'][i-1]
        while data['Problem Name'][i-subtractor] == prob and "".join([data['Anon Student Id'][i-subtractor], 
                                                                      data['Unit'][i-subtractor]]) == previous_idunit:
            validation.append(i-subtractor)
            subtractor += 1
    previous_idunit = current_idunit
    if i%500000 == 0:
        print i

subtractor = 1
prob = data['Problem Name'][i-1]
while data['Problem Name'][i-subtractor] == prob and "".join([data['Anon Student Id'][i-subtractor], 
                                                              data['Unit'][i-subtractor]]) == previous_idunit:
    validation.append(i-subtractor)
    subtractor += 1
```

```python
validation = sorted(validation)
validation_set = set(validation)
training = [i for i in range(len(data)) if i not in validation_set]
```

```python
validation = np.array(validation)
training = np.array(training)
```

```python
joblib.dump((training, validation), 'Pickles/cv_indices', compress=3)
```
