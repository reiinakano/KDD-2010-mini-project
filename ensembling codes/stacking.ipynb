### Neural networks

```python


def build_fn(no_hidden=3, hidden_activation='sigmoid', out_activation='sigmoid', 
             loss='mse', optimizer='rmsprop', random_state=0, 
             callbacks=[EarlyStopping(monitor='val_loss', patience=2)], 
             validation_split=0.2):
    
    np.random.seed(0)

    model = Sequential([
        Dense(no_hidden, input_dim=10),
        Activation(hidden_activation),
        Dense(1),
        Activation(out_activation),
    ])
    

    model.compile(loss=loss, optimizer=optimizer)
    return model

clf = KerasClassifier(build_fn)

param_grid = [
  {'batch_size': [128, 256], 'nb_epoch': [1000], 'no_hidden': [3, 6],
   'hidden_activation': ['relu', 'sigmoid', 'tanh'], 
   'optimizer': ['rmsprop', 'sgd', 'adam', 'adamax', 'nadam', 'adadelta' ,'adagrad'], 'verbose': [2], 
   'validation_split': [0.1],
   'callbacks': [[EarlyStopping(monitor='val_loss', patience=2)]]}
 ]

a = np.arange(487006)
np.random.shuffle(a)
my_cv = [(a[:460000], a[460000:])]

grid_search = GridSearchCV(clf, param_grid, cv=my_cv, 
                           scoring=basic_rms_scorer, n_jobs=1, verbose=10, refit=False)
```

```python
grid_search.fit(X, labels)
```

```python
def submit_file(probas, filename):
    ones = []
    zeros = []
    for zero_prob, one_prob in probas:
        if one_prob > zero_prob:
            ones.append(one_prob)
        else:
            zeros.append(one_prob)

    print len(ones), len(zeros)

    import csv
    with open(filename, 'wb') as csvfile:
        spamwriter = csv.writer(csvfile, delimiter='\t')
        spamwriter.writerow(['Row', 'Correct First Attempt'])
        for index, (zero_prob, one_prob) in enumerate(probas):
            spamwriter.writerow([index+1, float(one_prob)])
```
